train test split

```python
from sklearn.model_selection import train_test_split
```

```python
df_train , df_test =train_test_split(df,test_size=0.25)
```

preparing our data for modeling

```python
x_train = df_train.drop(columns=['target','target_name']).values
y_train=df_train['target'].values
```

```python
y_train
```

modeling- what is our baseline?

```python
#simple manual model

#tree based
def single_feature_prediction(petal_length):
  if petal_length < 2.5:
    return 0
  elif petal_length < 4.5:
    return 1
  else:
    return 2
```

```python
x_train[:,2]
```

```python
manual_y_prediction =np.array([single_feature_prediction(val) for val in x_train[:,2]])
```

```python
manual_model_accuracy =np.mean(manual_y_prediction == y_train)
```

```python
print(f"manual_model_accuracy :{manual_model_accuracy * 100:.2f}%")
```

modeling - logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

using a validation set to ealuate our model

```python
model=LogisticRegression()
```

```python
xt, xv, yt, yv = train_test_split(x_train, y_train, test_size=0.25)
```

```python
model.fit(xt, yt)
```

```python
y_pred =model.predict(xv)
```

```python
np.mean(y_pred == yv)
```

```python
model.score(xv,yv)
```

using cross-validation to evaluate our model

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies =cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy' )
```

```python
np.mean(accuracies)
```

where are we misclassification points?

```python
y_pred=cross_val_predict(model, x_train, y_train, cv=5)
```

```python
predicted_correctly_mask =y_pred == y_train
```

```python
not_predicted_correctly = ~predicted_correctly_mask
```

```python
x_train[not_predicted_correctly]
```

```python
df_predictions = df_train.copy()
```

```python
df_predictions['correct_prediction']= predicted_correctly_mask
```

```python
df_predictions['prediction']= y_pred
```

```python
df_predictions['correct_prediction']=df_predictions['prediction'].map({0 : 'setosa', 1 : 'versicolor' , 2 : 'virginica'})
```

```python
df_predictions.head()
```

```python
sns.scatterplot( x ='petal length (cm)',  y='petal width (cm)', hue='target_name' , data=df_predictions)
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs = axs.flatten()

    sns.scatterplot(x=x_axis_feature, y=y_axis_feature,
                   hue='prediction_label', data=df_predictions, ax=axs[0])
    axs[0].set_title('Model Predictions')

    sns.scatterplot(x=x_axis_feature, y=y_axis_feature,
                   hue='target_name', data=df_predictions, ax=axs[1])
    axs[1].set_title('True Labels')

    sns.scatterplot(x=x_axis_feature, y=y_axis_feature,
                   hue='correct_prediction', data=df_predictions, ax=axs[2])
    axs[2].set_title('Prediction Accuracy')

    axs[3].set_visible(False)

    plt.tight_layout()
    plt.show()
```

model tuning

```python
for reg_param in(0.1 ,0.3 , 0.9 , 1 , 1.3, 1.9):
  model = LogisticRegression(max_iter=200 , C=reg_param)
  accuracies =cross_val_score(model , x_train , y_train , cv=5 ,scoring='accuracy')
  print(f"accuracy :{np.mean(accuracies) * 100:.2f}%")
```
